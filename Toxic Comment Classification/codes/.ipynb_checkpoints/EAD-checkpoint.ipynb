### Distribution of Labels

```python
temp = df_train.loc[:, 'toxic' : 'clean'].sum(axis=0).reset_index(name = 'total')

fig = plt.figure(figsize = (9,6))
ax= sns.barplot(x = temp['index'], y = temp['total'], alpha = 0.8)
#adding the text labels
rects = ax.patches
labels = temp.total
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 5, label, ha='center', va='bottom')

plt.show()
```

### Correlation of Labels

```python
def cramers_corrected_stat(confusion_matrix):
    """ calculate Cramers V statistic for categorial-categorial association.
        uses correction from Bergsma and Wicher, 
        Journal of the Korean Statistical Society 42 (2013): 323-328
    """
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))    
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min( (kcorr-1), (rcorr-1)))
```

```python
# 'toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate'
col1="toxic"
col2="insult"
confusion_matrix = pd.crosstab(df_train[col1], df_train[col2])
print("Confusion matrix between toxic and severe toxic:")
print(confusion_matrix)
new_corr=cramers_corrected_stat(confusion_matrix)
print("The correlation between Toxic and Severe toxic using Cramer's stat=",new_corr)
```

## Feature Engineering
1. **Number of (Unique) Words**
2. **Number of Puncuations**
3. **Number of Sentences**
4. **Number of Cap Words**
5. **Mis-spelling**
6. **Number of curse word**

```python
# 1. number of words && number of unique words
df_train['word_count'] = df_train['comment_text'].apply(lambda x: len(x.lower().split(' ')))
df_train['uni_word_count'] = df_train['comment_text'].apply(lambda x: len(set(x.lower().split(' '))))

# 2. number of puncuations: ! or ?
punc_re = re.compile(r'[?!]+')
df_train['punc_counts'] = df_train['comment_text'].apply(lambda x: len(punc_re.findall(x)))

# 3. number of sentences
sentence_re = re.compile(r'[.!?]+[ ]')
df_train['sentence_count'] = df_train['comment_text'].apply(lambda x: len(sentence_re.split(x)))

# 4. number of Caps
cap_re = re.compile(r'\b[A-H-Z]+\b')
df_train['caps_count'] = df_train['comment_text'].apply(lambda x: len(cap_re.findall(x)))

# 5. mis-spelling ?????
```
